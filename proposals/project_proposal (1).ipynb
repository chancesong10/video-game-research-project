```python
import pandas as pd
import altair as alt
```

```python
players_data = pd.read_csv('players.csv')
players_data
```

```python
players_cleaned = players_data.drop(['hashedEmail', 'individualId', 'organizationName', 'name', 'experience', 'subscribe'], axis=1)
players_cleaned
```

```python
gender_plot = alt.Chart(players_cleaned).mark_bar().encode(
    x=alt.X('gender').title('Gender'),
    y=alt.Y('played_hours').title('Total Time Played (h)')
)

gender_plot
```
